```python
import pickle
import pandas as pd
```

```python
def load_config(f_name):
    with open( f_name, "rb" ) as f:
            config = pickle.load( f )
    return pd.DataFrame(config).transpose()
```

```python
config_4a = load_config('configs_4A.p')
config_4a.loc[config_4a['has_hidden_p'] == False, 'hidden_p'] = 0
config_4a.loc[config_4a['has_weight_p'] == False, 'weight_p'] = 0
config_4a = config_4a.drop(columns=['has_hidden_p', 'has_weight_p'])
config_4a.index.names = ['Dataset','Model Structure','Model Size']
config_4a
```

```python
print(config_4a.to_latex(index=True, float_format="%.2e", column_format='llllllllll'))
```

```python
config_4b = load_config('configs_4B.p')
config_4b.index.names = ['Dataset','Model Structure','Model Size']
config_4b
```

```python
print(config_4b.to_latex(index=True, float_format="%.2f", column_format='llllllllll'))
```
